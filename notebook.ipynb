```python
genre_to_emotion = {
    'Action': ['Excited'],
    'Adventure': ['Happy', 'Excited'],
    'Animation': ['Happy', 'Relaxed'],
    'Biography': ['Inspired', 'Sad'],
    'Comedy': ['Happy', 'Relaxed'],
    'Crime': ['Scared', 'Excited'],
    'Documentary': ['Inspired'],
    'Drama': ['Sad', 'Romantic'],
    'Family': ['Happy', 'Relaxed'],
    'Fantasy': ['Excited', 'Happy'],
    'Film-Noir': ['Scared'],
    'History': ['Inspired', 'Sad'],
    'Horror': ['Scared'],
    'Music': ['Happy', 'Inspired'],
    'Musical': ['Happy', 'Relaxed'],
    'Mystery': ['Scared', 'Excited'],
    'Romance': ['Romantic', 'Happy'],
    'Sci-Fi': ['Excited', 'Sad'],
    'Sport': ['Inspired'],
    'Thriller': ['Excited', 'Scared'],
    'War': ['Sad', 'Inspired'],
    'Western': ['Excited']
}
```

```python
def map_genres_to_emotions(genres):
    emotions = set()  # Usamos un set para evitar duplicados
    for genre in genres:
        if genre.strip() in genre_to_emotion:
            emotions.update(genre_to_emotion[genre.strip()])  # Agrega las emociones correspondientes al género
    return list(emotions)  # Convertir de nuevo a lista
```

```python
df['emotions'] = df['genre'].apply(map_genres_to_emotions)
```
